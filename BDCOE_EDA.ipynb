# `Spotify Features EDA`

- Here we have a datset that has different features of songs on spotify like popularity,acousticness, danceability,energy,etc.
- I have performed EDA on this dataset and I have driven out some insights from this data.
- Please find all the plots and insights below.

```python
#Importing all the necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the dataset
df = pd.read_csv('SpotifyFeatures.csv')
```

```python
# Getting the number of rows and columns in the datset
df.shape
```

```python
#Getting the mean,median,min,max and standard deviation of all the numerical data
df.describe()
```

### `Checking for missing values`

```python
#Getting the number of missing values column-wise
df.isna().sum()
```

- Since there is only one value missing in such a big data, so we should drop that row as it won't have a very big effect on the dataset

```python
#Dropping the row with missing value
df = df.dropna(subset = ['track_name'])
```

```python
#Checking if any other missing value remains
df.isna().sum()
```

```python
#Checking any duplicate values in the dataset
df.duplicated().sum()
```

```python
df.duplicated().any()
```

### `Comparing differnt song features with the popularity of the song`

```python
#Creating a scatter plot to compare how acousticness affect popularity of the song
x = df['popularity']
y = df['acousticness']
plt.figure(figsize=(20,20))
plt.scatter(x,y)
plt.title('Scatter plot to compare how acousticness affect popularity of the song',fontsize=20)
plt.xlabel('Popularity', fontsize=20)
plt.ylabel('Acousticness', fontsize=20)
plt.show()
```

- `As we can see that the scatter plot is not able to show us any insights as there is a lot of data that needs to be processed.`

```python
#Creating a hexbin to see if we get a clear view of the insights this way
x = df['popularity']
y = df['acousticness']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to compare how acousticness affect popularity of the song')
plt.xlabel('Popularity')
plt.ylabel('Acousticness')
plt.show()
```

- As we can see that songs with lower accousticness have higher popularity
- üí°Insight: We can say that song with lower accousticness have higher chances at getting popular.

```python
x = df['popularity']
y = df['danceability']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to show the relation between danceability and popularity')
plt.xlabel('Popularity')
plt.ylabel('Danceability')
plt.show()
```

- üí°Insight: Songs with higher danceability are more popular.

```python
x = df['popularity']
y = df['energy']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hex bin to show the relation between popularity and energy of the song')
plt.xlabel('Popularity')
plt.ylabel('Energy')
plt.show()
```

- üí°Insight: Songs with higher energy are more popular.

```python
x = df['popularity']
y = df['instrumentalness']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to show the relation between popularity and Instrumentalness of the song')
plt.xlabel('Popularity')
plt.ylabel('Instrumentalness')
plt.show()
```

- üí°Insight: Most popular songs don't have high instrumentalness

```python
x = df['popularity']
y = df['liveness']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to show the relation between popularity and liveness of the song')
plt.xlabel('Popularity')
plt.ylabel('Liveness')
plt.show()
```

- üí°Insight: Most popular songs don't have high liveness

```python
x = df['popularity']
y = df['loudness']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to show the relation between popularity and loudness of the song')
plt.xlabel('Popularity')
plt.ylabel('Loudness (in dB)')
plt.show()
```

- üí°Insight: Most popular songs have a loudness between -10 to 0 dB.

```python
x = df['popularity']
y = df['speechiness']*10
plt.hexbin(x,y,gridsize=20,cmap='plasma')
plt.title('Hexbin to show the relation between popularity and speechiness of the song')
plt.xlabel('Popularity')
plt.ylabel('Speechiness')
plt.show()
```

- üí°Insight: Most popular songs tend to have lower speechiness

```python
x = df['popularity']
y = df['valence']
plt.hexbin(x,y,gridsize=20,cmap='viridis')
plt.title('Hexbin to show the relation between popularity and Valence of the song')
plt.xlabel('Popularity')
plt.ylabel('Valence')
plt.show()
```

- üí°Insight: While valence might not be a key predictor of the popularity of the songs but songs with lower valence are most popular.

```python
#Creating a new column that stores the duration of the songs in seconds
df.loc[:,'duration_sec'] = df['duration_ms']/1000
```

```python
#Creating a scatter plot to check if duration plays any role in popularity of the song
df_filtered = df[df['duration_sec']<1000]
x = df_filtered['popularity']
y = df_filtered['duration_sec']
plt.title('Scatter plot to show the relation between popularity and duration of the song')
plt.scatter(x,y)
plt.xlabel('Popularity')
plt.ylabel('Duration of the track (in seconds)')
plt.show()
```

- As we can see that the scatter plot is not able to show us any insights. So we need to create a hexbin for this as well.

```python
x = df_filtered['popularity']
y = df_filtered['duration_sec']
plt.hexbin(x,y,gridsize=20)
plt.title('Hexbin in to show the relation between popularity and duration of the song')
plt.xlabel('Popularity')
plt.ylabel('Duration of the track (in seconds)')
plt.show()
```

- üí°Insights: Most popular songs have a duration of around 200-300 seconds.

```python
#Checking the total number of unique Artists in the dataset
df['artist_name'].nunique()
```

```python
#Seeing the top 10 artiss on spotify
top_artists_max = (
    df.groupby("artist_name")["popularity"]
      .max()
      .sort_values(ascending=False)
      .head(10)
)
print(top_artists_max)
```

```python
#Making a bar graph to show the number of songs every genre has
counts = df['genre'].value_counts()
label = counts.index
value = counts.values
plt.barh(label,value)
plt.title('Bargraph to show the number of songs (genre-wise)')
plt.xlabel('Number of songs')
plt.ylabel('Genre')
plt.show()
```

- üí°Insight: Most songs are from Comedy,Soundtrack,Indie and Pop genre

```python
#Bar graph to show the number of songs with according to the key used
counts = df['key'].value_counts()
label = counts.index
value = counts.values

plt.barh(label,value,color='c')
plt.xlabel('Number of songs')
plt.ylabel('Key')
plt.show()
```

- üí°Insight: Most songs use C,D and G as their primary key.

```python
genres = df["genre"].unique()
data = [df[df["genre"] == g]["popularity"].values for g in genres]

plt.figure(figsize=(14,6))
plt.boxplot(data, tick_labels=genres, patch_artist=True)

plt.xlabel("Genre")
plt.ylabel("Popularity")
plt.xticks(rotation=90)
plt.show()
```

```python
keys = df["key"].unique()

data = [df[df["key"] == k]["popularity"].values for k in keys]

plt.figure(figsize=(12,6))
plt.boxplot(data, tick_labels=keys, patch_artist=True)

plt.xlabel("Key")
plt.ylabel("Popularity")
plt.xticks(rotation=0)
plt.show()
```

```python
df_duration = df.groupby('genre')['duration_sec'].mean()
x = df_duration.index
y = df_duration.values
plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

```python
df_duration = df.groupby('genre')['duration_sec'].max()
x = df_duration.index
y = df_duration.values
bars = plt.barh(x,y,color='#fa6305')
for i,val in enumerate(y):
    if (val>4000):
        bars[i].set_hatch('/')
plt.xticks(rotation=90)
plt.show()
```
